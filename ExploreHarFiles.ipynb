```python
def k(o): return o.keys() if isinstance(o,dict) else len(o)
import datetime as dt
import re
def parse_timestamp( ts ) :
    conformant_ts = re.sub(r':(\d\d)$', r'\1', ts)
    return dt.datetime.strptime(conformant_ts,'%Y-%m-%dT%H:%M:%S.%f%z')
```
