```python
import sys
import pandas as pd

import numpy as np
import matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

sys.path.insert(0, 'modules/')
from AboutSPDY.har_bundle import har_to_pandas, datetimediff_to_microseconds

frame_1 = har_to_pandas( open ("data/ipython.zunzun.se.--ffspdy--.har") )
frame_2 = har_to_pandas( open ("data/mail.google.com.--ffspdy--.har") )
frame_3 = har_to_pandas( open ("data/www.facebook.com.--ffspdy--.har"))

section_1_data = [frame_1, frame_2, frame_3]
section_1_legends = ['ipython', 'gmail', 'facebook']
```

```python
class FrameStatistics(object):
    def __init__(self, frame):
        self.frame = frame 
    
    @property
    def body_size(self):
        f = self.frame
        s = f['sz_response_body'].sum()
        return s
    
    @property
    def response_headers_size(self):
        f = self.frame
        s = f['sz_response_headers'].sum()
        return s
        
    @property
    def download_total(self):
        f = self.frame
        s = f['sz_response_body'].sum()
        s += f['sz_response_headers'].sum()
        return s
```

```python
section_1_statistics = map(FrameStatistics, section_1_data)
df_stats = pd.DataFrame([ { 'total': s.download_total,
                 'body': s.body_size,
                 'headers': s.response_headers_size
                 }for s in section_1_statistics])
df_stats
```

```python
percent_headers = df_stats['headers']/df_stats['total']
percent_headers.name = '%headers'
percent_body = df_stats['body']/df_stats['total']
percent_body.name = '%body'
df_stats = pd.concat([
    df_stats,
    percent_headers,
    percent_body
    ],
    axis = 1
)
```

```python
fig_1 = plt.figure(figsize=(10.5,3))
radi = df_stats['total'] / df_stats['total'].max()
for i in range(3):
    ax = fig_1.add_subplot(131+i)
    ax.pie(df_stats.loc[i][:2], radius=radi[i])
    ax.set_xlabel(section_1_legends[i])
```

Everything is blue, so most of the data is contained on the bodies, not on the headers. So far at least.
